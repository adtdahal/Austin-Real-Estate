# 03. Correlation Matrix Heatmap

```python
# Creating a subset w/out latiude, longitude, yearBuilt, saleMonth, saleYear

sub = df_2[['propertyTaxRate', 'price', 'numPriceChanges', 'numOfPhotos', 'numOfAccessibilityFeatures', 'numOfAppliances', 'numOfParkingFeatures', 'numOfPatioAndPorchFeatures', 'numOfSecurityFeatures', 'numOfWaterfrontFeatures', 'numOfWindowFeatures', 'numOfCommunityFeatures', 'lotSizeSqFt', 'livingAreaSqFt', 'numOfPrimarySchools', 'numOfElementarySchools', 'numOfMiddleSchools', 'numOfHighSchools', 'avgSchoolDistance', 'avgSchoolRating', 'avgSchoolSize', 'medianStudentsPerTeacher', 'numOfBathrooms', 'numOfBedrooms', 'numOfStories', 'age']]
```

```python
sub
```

```python
# Creating a subplot w/ matplotlib

f,ax = plt.subplots(figsize=(20,20))

# Creating correlation heatmap in seaborn by applying a heatmap onto the correlation matrix & the subplots defined above
# 'annot' arguments allows plot to place the correlation coefficients onto the map

corr = sns.heatmap(sub.corr(), annot = True, ax = ax)
```

The variable I am most interested in is price, which had a weak correlation with majority of the variables. These are the top 5 variables listed by coefficient strength:

● livingAreaSqFt  (0.63) - Strong   
● numOfBathrooms  (0.55) - Strong        
● numOfBedrooms   (0.33) - Moderate      
● avgSchoolRating (0.30) - Weak/Moderate   
● medianStudentsPerTeacher (0.21) - Weak     

I was originally surprised that lotSizeSqFt had a weak correlation with price. After doing research, it turns out Zillow lists the overall lot size for condos and apartments, which is not necessarily representative.

# 06. Categorical Plot

```python
# Checking headers

df_2.head()
```

```python
# Creating histogram to visualize 'price' distribution

sns.histplot(df_2['price'], bins = 50, kde = True)
```

```python
# Calculating median prices by year

df_2.groupby('saleYear').agg({'price':['median']})
```

```python
# Creating new 'value' variable to categorize which homes were above/below median based on the year it was sold

df_2.loc[(df_2['price']<385990) & (df_2['saleYear']==2018), 'value'] = 'Below Median'
df_2.loc[(df_2['price']>=385990) & (df_2['saleYear']==2018), 'value'] = 'Median or Above'
df_2.loc[(df_2['price']<395315.5) & (df_2['saleYear']==2019), 'value'] = 'Below Median'
df_2.loc[(df_2['price']>=395315.5) & (df_2['saleYear']==2019), 'value'] = 'Median or Above'
df_2.loc[(df_2['price']<434000) & (df_2['saleYear']==2020), 'value'] = 'Below Median'
df_2.loc[(df_2['price']>=434000) & (df_2['saleYear']==2020), 'value'] = 'Median or Above'
df_2.loc[(df_2['price']<520000) & (df_2['saleYear']==2021), 'value'] = 'Below Median'
df_2.loc[(df_2['price']>=520000) & (df_2['saleYear']==2021), 'value'] = 'Median or Above'
```

```python
# Checking values

df_2['value'].value_counts(dropna = False)
```

```python
# Creating categorical plot - livingAreaSqft vs. price

sns.set(style="ticks")
sns.scatterplot(x = 'livingAreaSqFt', y='price', hue='value', data=df_2)
```

Majority of homes valued below median price have living areas no bigger than 6,000 sqft. Homes valued above median appear to have a wider price range.

```python
# Creating categorical plot - avgSchoolRating vs. price

sns.set(style="ticks")
sns.scatterplot(x = 'avgSchoolRating', y='price', hue='value', data=df_2)
```

Homes below and above the median have ratings all across the scale. However, schools rated 9+ are made up exclusively of median/above median valued homes. These schools are most likely in well-to-do areas and have the funds to provide good education and programs.

```python
# Creating new ageCategory column

df_2.loc[df_2['age']<=10,'ageCategory'] = '<=10 Yrs'
df_2.loc[(df_2['age']>10) & (df_2['age']<=20), 'ageCategory'] = '10-20 Yrs'
df_2.loc[(df_2['age']>20) & (df_2['age']<=30), 'ageCategory'] = '20-30 Yrs'
df_2.loc[(df_2['age']>30) & (df_2['age']<=40), 'ageCategory'] = '30-40 Yrs'
df_2.loc[(df_2['age']>40) & (df_2['age']<=50), 'ageCategory'] = '40-50 Yrs'
df_2.loc[df_2['age']>50, 'ageCategory'] = '50+ Yrs'
```

```python
# Checking values

df_2['ageCategory'].value_counts(dropna = False)
```

```python
# Creating categorical plot - avgSchoolRating vs. price by ageCategory

sns.set(style="ticks")
sns.scatterplot(x = 'livingAreaSqFt', y='price', hue='ageCategory', data=df_2)
```

Homes in different age categories have a wide range of living area square feet (2,000-7,000). Once the square feet is past 7,000 we mainly see homes that are no older than 30 years.
