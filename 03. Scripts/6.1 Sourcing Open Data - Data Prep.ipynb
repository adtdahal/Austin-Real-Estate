###### Some of the numbers listed are raising some questions. The first one being the max amount of garage space being 22?

```python
# Checking all values for garageSpaces column

df_house_3['garageSpaces'].value_counts(dropna = False)
```

###### I checked the original data source, zillow.com, and it appears the web scraper took the values in parkingSpaces for garageSpaces as well since the garageSpaces variable is not explicitly stated. parkingSpaces is a broad term and can entail different definitions which are not clearly explained. The column will be dropped due to the possibility of inaccurate data.

```python
# Dropping garageSpaces column

df_house_4 = df_house_3.drop(columns = ['garageSpaces'])
```

```python
df_house_4.head(1)
```

###### numOfBathrooms maximum is 27, this is a bit odd

```python
# Checking numOfBathrooms values

df_house_4['numOfBathrooms'].value_counts(dropna = False)
```

###### Bathroom values can end in .0, .5, or .75. I am assuming 1.70 was a mistake and will replace it with 2 since it is a more common value. 

```python
# Replacing 1.70 value with 2.00

df_house_4.loc[df_house_4['numOfBathrooms'] == 1.70, 'numOfBathrooms'] = 2.00
```

```python
# Checking numOfBathrooms values

df_house_4['numOfBathrooms'].value_counts(dropna = False)
```

###### 27 may have been a typo for either: 2.75, 2.50, or 3.00. I will go with 3 since it is more common compared to the other two values

```python
# Replacing 27 value with 3

df_house_4.loc[df_house_4['numOfBathrooms'] == 27.00, 'numOfBathrooms'] = 3.00
```

###### There are 125 '0' values, this needs to be explored further

```python
# Looking for rows with 0 bathrooms listed

df_house_4.loc[df_house_4['numOfBathrooms'] == 0]
```

###### It appears a lot of the 0 bathrooms belong to Vacant Land property type

```python
# Checking homeType values

df_house_4['homeType'].value_counts(dropna = False)
```

```python
# Creating new df without 'Vacant Land' & 'Other' - only interested in land being used; unsure what 'Other' would consist of

df_house_5 = df_house_4[(df_house_4['homeType']!='Vacant Land') & (df_house_4['homeType']!='Other')]
```

```python
# Double checking homeType values

df_house_5['homeType'].value_counts(dropna = False)
```

```python
# Re-checking rows with 0 bathrooms listed

df_house_5.loc[df_house_5['numOfBathrooms'] == 0]
```

```python
# Viewing in Excel

df_house_5.loc[df_house_5['numOfBathrooms'] == 0].to_clipboard()
```

###### Only 48 rows contain 0 bathrooms. I am assuming the data was unavailable or missing on Zillow. These rows will be removed since they make up less than 1% of total data

```python
# Creating new df w/ bathroom values

df_house_6 = df_house_5.loc[df_house_5['numOfBathrooms'] > 0]
```

```python
df_house_6.shape
```

```python
# Checking numOfBedroom values - original descriptive analysis showed minimum was 0

df_house_6['numOfBedrooms'].value_counts(dropna = False)
```

```python
# Taking a closer look into 0 numOfBedroom values

df_house_6.loc[df_house_5['numOfBedrooms'] == 0]
```

###### Only 10 rows contain 0 bedrooms, will remove these as well

```python
# Creating new df w/ bedroom values

df_house_7 = df_house_6.loc[df_house_6['numOfBedrooms'] > 0]
```

```python
df_house_7.shape
```

```python
# Checking bedroom maximum value of 20

df_house_7.loc[df_house_7['numOfBedrooms'] == 20]
```

###### The squarefoot seems too small for there to be 20 bedrooms. It also does not make sense there would only be 2 bathrooms for 20 bedrooms. I will assume this is a typo and will replace 20 with 2.

```python
# Replacing bedroom value of 20 with 2

df_house_7.loc[df_house_7['numOfBedrooms'] == 20, 'numOfBedrooms'] = 2
```
